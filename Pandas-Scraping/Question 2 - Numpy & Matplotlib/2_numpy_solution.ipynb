# Question 2: Numpy & Matplotlib 

For this question you use the dataset Swim_Sep 25.csv.  The dataset contains for September 25 all laps swum by a swimmer and has following columns:
* Laps: number of lap
* Total laptime: time to swim a lap plus the turn time
* Turn time: time the swimmer takes to turn
* Swim time
* Strokes: number of strokes needed for the lap
* Rest: rest time

**Step 1:**
Import the necessary libraries Numpy & Matplotlib.

```python

#Step 1:

import matplotlib.pyplot as plt
import numpy as np
```

**Step 2:** 
* Load the data from Swim_Sep 25.csv in a numpy array of datatype float. 
* Skip the first row which contains the column headers.  
* Print the first 5 rows to check the data are loaded.

```python
#Step 2:
swimData = np.loadtxt(fname='./Swim_Sep 25.csv',   delimiter=',',dtype=float, skiprows=1)
print(swimData[:5])
```

**Step 3:** Give an overview of the shortest lap, the longest lap and the total swim time over all the laps.

Make sure it looks like this:

![](../files/Step3_numpy.png)

```python
#Step 3

shortlapArray = np.min(swimData[:,1])
longestlapArray = np.max(swimData[:,1])
totalSwimTime = np.sum(swimData[:,3])
print(f"Shortest lap: {shortlapArray}\n"
      f"Longest lap: {longestlapArray}\n"
      f"Total swimtime: {totalSwimTime}")
```

**Step 4:** Calculate for the variables 'Total Laptime'  and 'Turn Time' the overall total, the mean and the standard deviation. Round the numbers to 3 decimal places.

![](../files/Step4_numpy.png)

```python
#Step 4

laptimeArray = swimData[:,1]
laptimeTotal = np.sum(laptimeArray).round(3)
laptimeMean = np.mean(laptimeArray).round(3)
laptimeSD = np.std(laptimeArray).round(3)

turntimeArray = swimData[:,2]
turntimeTotal = np.sum(turntimeArray).round(3)
turnimeMean = np.mean(turntimeArray).round(3)
turntimeSD = np.std(turntimeArray).round(3)

print(f"Total laptime {laptimeTotal} {laptimeMean} {laptimeSD}\n"
      f"Turn Time {turntimeTotal} {turnimeMean} {turntimeSD}")
```

Create a chart that plots the frequency distribution for total laptime.

**Step 6:**   
Plot the correlation between the number of strokes and the swim time in seconds.  
The plot looks like this:  

![](../files/strokesswimtime.PNG)

```python
# Step 6

strokesArray = swimData[:,4]
swimArray = swimData[:,3]
plt.scatter(strokesArray, swimArray, color='blue', marker='o')
plt.xticks(np.arange(0, 40, step=5))
plt.xlabel('Number of strokes')
plt.ylabel('Swim time in seconds')
plt.show()
```
