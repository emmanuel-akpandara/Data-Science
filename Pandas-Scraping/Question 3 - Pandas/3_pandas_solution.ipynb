For this question you work with 2 csv files containing data on car sales: 
  1. Sales.csv contains general information about the sale of the car like manufacturer, model, price, ...  
  2. Customers.csv contains information about the customer like gender, age, annual income, profession, ... 
 


**Step 1:**
* Import the necessary libraries.  
* Load the two files "Sales.csv" and "Customers.csv" in 2 separate dataframes. 
* Print the first 5 records of each dataframe.

```python
# Step 1:
import pandas as pd
salesFile = pd.read_csv('./Sales.csv',sep=',',index_col="CustomerID")
CustomerFile= pd.read_csv('./Customers.csv',sep=',',index_col="CustomerID")
firstFiveSales = salesFile[:5]
firstFiveCustomer = CustomerFile[:5]
print(firstFiveSales)
print(firstFiveCustomer)
```

 **Step 2:**   
 * Combine the 2 dataframes into 1 dataframe named 'df_sales'. Make sure you keep all the rows in Customers.csv!
 * Print the first 5 rows of the combined dataframe.
 * Print an overview of all the columns in the combined dataframe. 


```python
# Step 2:
df_sales = pd.merge(salesFile,CustomerFile,left_index=True,right_on='CustomerID')
print(df_sales[:5]) #first 5 rows
print(df_sales.info) #overview

```

 **Step 3:**
